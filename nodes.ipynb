```python
class node(object):
    # Constructor; build node
    def __init__(self,name,value):
        self.value = value
        self.name = name
        self.neighbors = []
        self.cache = 0 # place to hold temp values
        return None
    # Report neighbors for given node
    def get_neighbors(self):
        return self.neighbors
    # Calculate where chips go
    def transfer(self):
        if self.value >= len(self.neighbors):
            # How many do I give to each
            transfer = int((self.value - (self.value % len(self.neighbors)))/len(self.neighbors))
            self.value = self.value % len(self.neighbors)
            for n in self.neighbors:
                n.cache += transfer
            return True
        else:
            return False
    def add(self,val):
        self.value += val
    def update(self):
        self.add(self.cache)
        self.cache = 0
```

```python
graphs = []
graphs.append(node('a',5))
graphs.append(node('b',2))
graphs.append(node('c',1))
graphs.append(node('d',0))
```

```python
graphs[0].neighbors.append(graphs[1])
graphs[0].neighbors.append(graphs[2])
graphs[1].neighbors.append(graphs[0])
graphs[1].neighbors.append(graphs[3])
graphs[2].neighbors.append(graphs[0])
graphs[2].neighbors.append(graphs[3])
graphs[3].neighbors.append(graphs[1])
graphs[3].neighbors.append(graphs[2])
```

```python
def report(graph):
    print([x.value for x in graph])

for y in graphs:
    print(y.name,y.value, "neighbors:")
    for x in y.neighbors:
        print("\t",x.name)
```

```python
for _ in range(4000):
    for x in graphs:
        x.transfer()
    for x in graphs:
        x.update()
    report(graphs)
```
